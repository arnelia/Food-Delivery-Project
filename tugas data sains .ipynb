BASIC PROBLEM DAN ALGORITMA

Dimisalkan May memiliki dataset berupa food delivery, yanga kan digunakan untuk mengukur frekuensi order dari Restoran P berdasarkan hari dan waktu. Algoritma untuk menentukannya adalah

1. Import dataset.
2. Gunakan kolom resto_id, date_time, dan status.
3. Ubah kolom date_time ke bentuk days dan time.
4. Cleaning data dengan status completed.
5. Visualisasi data

1. Import library and dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
dl = pd.read_csv('food_delivery_datasets.csv')
```

2. Declare the variables

```python
dl
```

```python
i = dl[['resto_id','status']]
i.head(5)
```

```python
dl['date_time'] = pd.to_datetime(dl['date_time']).dt.round('H')
dl['date_time']
```

```python
j = dl['date_time'].dt.strftime('%A')
j
```

```python
k = dl['date_time'].dt.strftime('%H:%M')
k
```

```python
dl = pd.concat([i['resto_id'], j, k, i['status']], axis=1)
dl.columns = ['resto_id', 'days', 'time', 'status']
dl.sample(10)
```

3. Clean the dataset with status completed only

```python
group_status = dl.groupby('status')
group_status['status'].size()
```

```python
dl = dl[dl.status == 'Completed']
dl.reset_index(drop=True)
```

4. Visualize the data

```python
# Pengelompokan data berdasarkan periode hari.

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plot1 = dl.groupby(['days']).size().reindex(days).plot(kind='bar', figsize = (10,6), ylabel='freq', title='Frequencies towards days of the period')
plot1.figure.savefig('../freq-over-days.jpg')
```

```python
dl.groupby(['days']).size().reindex(days)
```

Frekuensi order di Restoran P paling banyak terjadi pada weekend (Saturday dan Sunday)

```python
# # Pengelompokan data berdasarkan periode waktu. (kumulatif)

fig, ax = plt.subplots()
hours = mdates.HourLocator(interval = 1)

hm = dl.groupby(['time']).size()

ax.plot(hm.index, hm.values)
hm.plot(ax = ax, x_compat=True, figsize = (10,6), ylabel='freq', 
        title='Frequencies towards times of the period')
ax.xaxis.set_major_locator(hours)

fig.autofmt_xdate()
fig.savefig('../freq-over-cumtimes.jpg')
```

```python
hm = dl.groupby(['time']).size()
hm.head(10)
```

Pengguna tidak dapat membuat order makanan di Restoran P pada pukul 02:00-06:00, karena pada dini hari server berada dalam kondisi downtime.

```python
# Pengelompokan data berdasarkan periode waktu hari.

group_days = dl.groupby(['days', 'time']).size().unstack(level=0, fill_value=0)
plot3 = group_days.plot(figsize=(10,6), ylabel='freq',
                title='Multigraph of frequencies towards times of the period')
                
plot3.figure.savefig('../freq-over-times-multi.jpg')
```

```python
dl.groupby(['days', 'time']).size().idxmax()
```

Frekuensi order paling banyak terjadi saat jam makan siang (antara jam 12:00-14:30) dan jam pulang kantor (antara jam 18:00-20:00)

```python
# Data visualization in 3d with frequencies as z.

fig = plt.figure('scatter dates')

ix = fig.add_subplot(111, projection='3d')

grp_days = dl.groupby(['days', 'time'], as_index=False).size()
time = grp_days['time'].unique()
time.sort()

grp_days['time'] = grp_days['time'].replace(time, [t for t in range(len(time))])
grp_days['days'] = grp_days['days'].replace(days, [t for t in range(len(days))])

x = grp_days['days']
y = grp_days['time']
z = grp_days['size']

plot = ax.scatter(x, y, z, k=z)
fig.colorbar(plot)

ix.xaxis.set_ticklabels(days)
ix.set_yticks([i for i in range(0, len(time), 3)])
ix.yaxis.set_ticklabels([time[t] for t in range(0, len(time), 3)])
plt.show()
```

```python
grp = dl.groupby(['days', 'time']).size()
print('Maximum frequencies occured in %s at %s for %i occurences' % (grp.idxmax()[0], grp.idxmax()[1], grp.max()))
```

Kesimpulan: frekuensi order makanan di Restoran P paling banyak pada hari minggu pukul 14:00, sebanyak 23 orderan.
